```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img=cv.imread('brain_proton_density_slice.png',cv.IMREAD_GRAYSCALE)
assert im is not None

m1=np.linspace(0,0,151).astype(np.uint8)
m2=np.linspace(255,255,20).astype(np.uint8)
m3=np.linspace(0,0,85).astype(np.uint8)
mp1=np.concatenate((m1,m2),axis=0).astype(np.uint8)
mp2=np.concatenate((mp1,m3),axis=0).astype(np.uint8)
transformed_image = mp2[img]

y_values = np.array(mp2).astype(np.uint8)
x_values = np.linspace(0,255,256).astype(np.uint8)

fig, ax = plt.subplots(1,3, figsize=(60,20))
ax[0].imshow(img, cmap='gray', vmin=0, vmax=255)
ax[0].xaxis.tick_top()
ax[0].set_title('Original', color = 'red',fontsize='40')
ax[1].imshow(transformed_image, cmap='gray', vmin=0, vmax=255)
ax[1].xaxis.tick_top()
ax[1].set_xlabel(r'Input Intensity',fontsize=25)
ax[1].set_ylabel('Output Intensity',fontsize=25)
ax[1].set_title('White Matter', color = 'blue',fontsize='40')
ax[2].plot(x_values,y_values, color = 'blue')
ax[2].set_xlabel(r'Original Intensity',fontsize=25)
ax[2].set_ylabel('Transformed Intensity',fontsize=25)
ax[2].set_title('Intensity Plot',fontsize='40')
plt.savefig('q2a_transform')
plt.show()

```
